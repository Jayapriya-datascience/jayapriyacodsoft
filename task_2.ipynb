```python
encoder=LabelEncoder()
new_col=encoder.fit_transform(df_combine["merchant"].values)
df_combine["merchant"]=new_col
df_combine.drop(labels=["merchant"],axis=1,inplace=True)
```

```python
encoder=LabelEncoder()
new_col1=encoder.fit_transform(df_combine["category"].values)
df_combine["category"]=new_col
df_combine.drop(labels=["category"],axis=1,inplace=True)
```

```python
df_combined=pd.get_dummies(df_combine)
df_combined.drop(labels=["gender_F"],axis=1,inplace=True)
```

```python
df_combine.head()
```

```python
x=df_combined.drop(labels=["is_fraud"],axis=1)
y=df_combined["is_fraud"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
lr_model=LogisticRegression()
lr_model.fit(X_train,y_train)
```

```python
lr_prediction=lr_model.predict(X_test)
```

```python
print(confusion_matrix(y_test,lr_prediction))
print("Logistic Regression Model:")
print(classification_report(y_test,lr_prediction))
print("Accuracy: {}".format(accuracy_score(y_test,lr_prediction)))
```

```python
dt_model=DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
```

```python
dt_predictions=dt_model.predict(X_test)
```

```python
print(confusion_matrix(y_test,dt_predictions))
print("Decision Tree Model:")
print(classification_report(y_test,dt_predictions))
print("Accuracy: {}".format(accuracy_score(y_test,dt_predictions)))
```

```python
rf_model=RandomForestClassifier()
```

```python
rf_model.fit(X_train, y_train)
rf_predictions=rf_model.predict(X_test)
```

```python
print("Random Forest Model:")
print(confusion_matrix(y_test,rf_predictions))
print(classification_report(y_test,rf_predictions))
print("Accuracy: {}".format(accuracy_score(y_test,rf_predictions)))
```
